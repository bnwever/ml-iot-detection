```python
import pandas as pd
```

```python
file_path = '../data/raw/ciciot2022.parquet'
df = pd.read_parquet(file_path)
```

```python
# Count and print devices by category
if 'device_name' in df.columns:
    counts = df['device_name'].value_counts()
elif 'label' in df.columns:
    counts = df['label'].value_counts()
else:
    counts = df['global_category'].value_counts()

print("--- Device/Category Distribution ---")
print(counts)

print("\n--- Percentage Distribution ---")
print(df['global_category'].value_counts(normalize=True) * 100)
```

```python
# Count and print devices by name
if 'device_name' in df.columns:
    counts = df['device_name'].value_counts()
elif 'label' in df.columns:
    counts = df['label'].value_counts()
else:
    counts = df['device'].value_counts()

print("--- Device/Category Distribution ---")
print(counts)

print("\n--- Percentage Distribution ---")
print(df['device'].value_counts(normalize=True) * 100)
```
