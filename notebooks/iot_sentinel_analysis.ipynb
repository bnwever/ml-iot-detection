```python
import pandas as pd
```

```python
file_path = '../data/processed/iot_sentinel.parquet'
df = pd.read_parquet(file_path)
```

```python
# Display the first 5 rows
df.head()
```

```python
# Count and print devices by name
counts = df['device_name'].value_counts()

print("--- Device Distribution ---")
print(counts)

print("\n--- Percentage Distribution ---")
print(df['device_name'].value_counts(normalize=True) * 100)
```
